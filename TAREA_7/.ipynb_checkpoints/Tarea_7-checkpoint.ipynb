# Solucion de la ecuacion de Burguer usando el método de la ecuación 2 propuesto en el enunciado de la tarea

```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_x = 200
n_t = 200
u_0=3

x = linspace(0,2, n_x)
 

dx =2*np.pi/(n_x-1)
beta=0.12
dt =dx*beta
```

```python
U= np.zeros(n_t*n_x).reshape(n_t, n_x)
```

```python
#Condicion Inicial
u_initial = u_0*sin(pi*x)

u_future = zeros(n_x)

u_past = u_initial.copy()
u_present = u_initial.copy()

U = np.zeros(n_t*n_x).reshape(n_t, n_x)
U[0]=u_past

for j in range(1,n_t):
    for i in range(1,n_x-1):
        u_future[i] = u_past[i] - (beta/2)*((u_present[i+1])**2-(u_present[i-1])**2)
    U[j] = u_future
    u_past = u_present.copy()
    u_present = u_future.copy()
```

```python
plot(x,U[1])
plt.show()
plt.close
plot(x,U[199])
plt.show()
plt.close
```

```python
x= linspace(0,2,n_x)
t = linspace(0,0.15,n_t)

fig = plt.figure(figsize(15,10))
ax = fig.gca(projection='3d')
x ,t = np.meshgrid(x,t)
surf = ax.plot_wireframe(x,t,U)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

plt.savefig("1.png")
plt.show()
```

# Solución de la Ecuación de Burguer con el segundo método     

```python
n_x = 200
n_t = 200
u_0=3

x = linspace(0,2, n_x)
u = ones(n_x)
un = ones(n_x) 

dx =2*np.pi/(n_x-1)
b=0.08
dt =dx*beta
#Condicion Inicial
u_initial = u_0*sin(pi*x)

u_future = zeros(n_x)

u_past = u_initial.copy()
u_present = u_initial.copy()

U = np.zeros(n_t*n_x).reshape(n_t, n_x)
U[0]=u_past

for j in range(1,n_t):
    for i in range(1,n_x-1):
           u_future[i] = u_present[i] - b/4*((u_present[i+1])**2-(u_present[i-1])**2)\
            + b**2/8 *((u_present[i+1]+u_present[i])*((u_present[i+1])**2-(u_present[i-1])**2)\
            - (u_present[i]+u_present[i-1])*((u_present[i+1])**2-(u_present[i-1])**2))
    U[j] = u_future
    u_past = u_present.copy()
    u_present = u_future.copy()
    
plot(x,U[2])
plt.show()
plt.close
plot(x,U[199])
plt.show()
plt.close
```
